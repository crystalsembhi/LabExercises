```python
from tkinter import *
import requests
import json
import random
```

```python
root = Tk()
root.title("Quotes")
root.geometry("650x500")

value = StringVar()
value.set("Select an author")


def get_author():
    num = 5
    global authorquotelist
    authorquotelist = {}
    while num != 0:
        url = requests.get("https://api.quotable.io/random")
        file = url.json()
        global author
        author = file["author"]
        quote = file["content"]
        authorquotelist[author] = quote
        num -= 1
           
    
def quote():
    quotebox = Text(root, bg="lightgray", padx=50, pady=50)
    quotebox.grid(row=4, column=1)
    get_author()
    author = value.get()
    for i,j in authorquotelist.items(): #https://www.plus2net.com/python/tkinter-OptionMenu.php
        if i==author: 
            global actualquote
            actualquote = j 
    quotebox.insert(END, actualquote)

          

def reload():
    num = 5
    global authorquotelist
    authorquotelist = {}
    while num != 0:
        url = requests.get("https://api.quotable.io/random")
        file = url.json()
        global author
        author = file["author"]
        quote = file["content"]
        authorquotelist[author] = quote
        num -= 1
    dropdown = OptionMenu(root, value, *authorquotelist.keys())
    dropdown.grid(row=1, column=1)




dropdown = OptionMenu(root, value, *authorquotelist.keys())
dropdown.grid(row=1, column=1)

choose_button = Button(root, text="Click to choose", command=quote)
choose_button.grid(row=2, column=1)

reload_button = Button(root, text="Reload Menu", command=reload) 
reload_button.grid(row=3, column=1)
#the menu doesn't go back to "Select an Author", but it does refresh with new options
                
root.mainloop()
```

### Weather GUI

```python
root = tk.Tk()
root.title('Weather Forecast')
root.geometry('600x210')

columns = ('Day', 'Temperature', 'Precipitation')

tree = ttk.Treeview(root, columns=columns, show='headings') #treeview widget

tree.heading('Day', text='Day')
tree.heading('Temperature', text='Temperature (F)')
tree.heading('Precipitation', text='% Precipitation')


def getting_weather():
    lat = "42.098701"
    lon = "-75.912537"
    genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")
    json_file = genius.json()

    forecast = json_file["properties"]["forecast"]
    new_request = requests.get(forecast)

    new = new_request.json()
    new1 = new["properties"]["periods"]
    
def info():
    for i in range(len(new1)):
        getting_weather()
        global result
        result = []
        result.append(new1[i]["name"])
        result.append(new1[i]["temperature"])
        result.append(new1[i]["probabilityOfPrecipitation"]["value"])
        tree.insert('', "end", values=(result))

info()
        
tree.grid(row=0, column=0, sticky='nsew')


root.mainloop()

#Sources: 
#https://www.pythontutorial.net/tkinter/tkinter-treeview/
#https://www.plus2net.com/python/tkinter-treeview-insert.php
```
