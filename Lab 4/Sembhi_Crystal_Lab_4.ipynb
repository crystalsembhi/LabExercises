```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd
```

```python
csv_file = open("imbd.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["title", "year", "genre"])

source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")
section = soup.find("div", class_="lister-item-content")

for element in soup.find_all("div", class_="lister-item-content"):

    title = element.a.text.strip()
    year = element.find("span", class_="lister-item-year text-muted unbold").text.strip()
    genre = element.find("span", class_="genre").text.strip()
    print(title)
    print(year)
    print(genre)
    csv_writer.writerow([title, year, genre])
    
csv_file.close()
```

```python
csv_file = open("cats.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Title", "Description", "Link", "Contributor", "Date"])

page = 1 

while page != 6:

    source2 = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup2 = BeautifulSoup(source2, "lxml")

    for element in soup2.find_all("div", class_="description"):
        
        title2 = element.a.text.strip()
        
        if element.find('span',{'class':'item-description-abstract'}):
            description = element.find('span',{'class':'item-description-abstract'}).text.strip()
      
        link = element.a.get("href")
    
        if element.find('li',{'class':'contributor'}):
            contributor = element.find('li',{'class':'contributor'}).span.text.strip()
            
        if element.find('li',{'class':'date'}):
            date = element.find('li',{'class':'date'}).span.text.strip(": ")
         
        print(title2)
        print(link)
        print(description)
        print(contributor)
        print(date)
        
        print("")

        csv_writer.writerow([title2, description, link, contributor, date])       
       
    page +=1

#Used source listed below to get if statement for description, contributor, and date information
```
